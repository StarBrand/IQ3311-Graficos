# Ley de los Gases Ideales

```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
n = 1
R = 0.08205746
```

```python
FOLDER = "gases_ideales"
os.makedirs(FOLDER, exist_ok=True)
```

```python
def plot_law(x, y_func, constants, x_label, y_label, const_label, title):
    fig, ax = plt.subplots(figsize=(6, 6))
    y_min = float("inf")
    y_max = -float("inf")
    for const in constants:
        y = y_func(x, const)
        y_min = min(y_min, y.min())
        y_max = max(y_max, y.max())
        ax.plot(x, y, label=const_label.format(const))
    ax.set_xlabel(x_label, fontsize=14)
    ax.set_ylabel(y_label, fontsize=14)
    ax.set_xlim((x[0], x[-1]))
    ax.set_ylim((y_min, y_max))
    ax.set_title(title + "\n", fontsize=18)
    ax.grid()
    ax.legend(fontsize=14)
    return fig, ax
```

## Ley de Boyle-Mariotte

```python
def pressure(volume, temperature):
    return n * R * temperature / volume
```

```python
temperature = [250, 350, 450]
volume = np.linspace(10, 60)

fig, ax = plot_law(volume, pressure, temperature, "V [L]", "P [atm]", "T = {:02} [K]", "Ley de Boyle-Mariotte")
fig.savefig(os.path.join(FOLDER, "ley_boyle_mariotte.png"))
```

## Ley de Charles

```python
def volume(temperature, pressure):
    return n * R * temperature / pressure
```

```python
pressure = [0.5, 1.0, 3.0]
temperature = np.linspace(0, 450)

fig, ax = plot_law(temperature, volume, pressure, "T [K]", "V [L]", "P = {:02} [atm]", "Ley de Charles")
fig.savefig(os.path.join(FOLDER, "ley_charles.png"))
```

## Ley de Gay-Lussac

```python
def pressure(temperature, volume):
    return n * R * temperature / volume
```

```python
volume = [10, 20, 60]
temperature = np.linspace(0, 450)

fig, ax = plot_law(temperature, pressure, volume, "T [K]", "P [atm]", "V = {:02} [L]", "Ley de Gay-Lussac")
fig.savefig(os.path.join(FOLDER, "ley_gay_lussac.png"))
```

## Ley de los Gases Ideales

```python
from mpl_toolkits.mplot3d import axes3d
```

```python
volume = np.linspace(10, 60)
temperature = np.linspace(0, 450)

V, T = np.meshgrid(volume, temperature)

P = n * R * T / V
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(V, T, P, rstride=5, cstride=1, cmap='coolwarm', edgecolor='none')
ax.set_xlabel("V [L]", fontsize=14)
ax.set_ylabel("T [K]", fontsize=14)
ax.set_zlabel("P [atm]", fontsize=14)
ax.set_title("Ley de los Gases Ideales", fontsize=18)
ax.grid()

fig.savefig(os.path.join(FOLDER, "ley_gases.png"))
```
