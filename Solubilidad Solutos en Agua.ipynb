# Solubilidad de diferentes solutos

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PLOT_PATH = "solubilidad"

os.makedirs(PLOT_PATH, exist_ok=True)
```

```python
sol_nacl = 35.6
sol_kno3 = pd.read_csv(os.path.join("data", "sol_kno3.csv"), sep=",", header=0)
sol_na2so4 = pd.read_csv(os.path.join("data", "sol_na2so4.csv"), sep=",", header=0)
```

```python
sol_na2so4.loc[len(sol_na2so4), "T[°C]"] = 100
sol_na2so4.loc[len(sol_na2so4) - 1, "Solubility"] = sol_nacl
```

```python
sol_nacl = pd.DataFrame({
    "T[°C]": [0, 100],
    "Solubility": [sol_nacl] * 2
})
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
for marker, name, data in [("o", "NaCl", sol_nacl), ("*", "KNO₃", sol_kno3), ("^", "Na₂SO₄", sol_na2so4)]:
    ax.plot(data["T[°C]"], data["Solubility"], marker=marker, label=name, markersize=10, linewidth=3.0)
ax.set_xlabel("T[°C]", fontsize=14)
ax.set_ylabel("Solubilidad[g/100 g H₂O", fontsize=14)
ax.set_xlim(0, 100)
ax.set_title("Solubilidad de solutos en agua\n", fontsize=18)
ax.grid()
ax.legend(fontsize=14)

fig.savefig(os.path.join(PLOT_PATH, "solubilidad_solutos.png"))
```
